#### 4.1 Distribution of Calories
The histogram and boxplot for the said distribution can be plotted together with the help of a subplot, so that we can compare both the visuals side by side to get more comprehensive insights about the distribution as well as the outliers.



```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['Calories'], bins=10, kde=True)
plt.title('Histogram of Calorie Distribution')
plt.xlabel('Calories')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.boxplot(x=df['Calories'])
plt.title('Box Plot of Calorie Distribution')
plt.xlabel('Calories')

plt.tight_layout()
plt.show()
```

The histogram reveals a right-skewed distribution, with a peak between 125 and 500 calories, and a long tail extending toward higher calorie values. Most menu items cluster on the lower end of the calorie spectrum, while only a few high-calorie outliers are present. The box plot supports this observation, showing a median calorie count of approximately 340 calories, with the 25th and 75th percentiles at 210 and 500 calories, respectively. The plot also highlights several outliers, particularly high-calorie items, beginning with the "McFlurry with Reese's Peanut Butter Cups (Medium)" at 810 calories.
